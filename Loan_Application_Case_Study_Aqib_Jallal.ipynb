## Univariate Analysis on Categorical Variable

```python
# Creating method to find the percentage of defaulters
def percentage_Defaulters(column_1, column_2, xlabel):
    application_data_temp = application_data[[column_1,column_2]].groupby([column_2], as_index=False).mean()
    application_data_temp[column_1] = application_data_temp[column_1]*100
    application_data_temp.sort_values(by=column_1, ascending=False, inplace=True)
    sns.barplot(x=column_2, y = column_1, data = application_data_temp)
    plt.xlabel(xlabel)
    plt.ylabel('Defaulter %')
    #plt.xticks(rotation=45, ha='right')
```

```python
#Analysing defaulters in Contract Type
plt.figure(figsize = (20,20))
plt.subplot(2,2,1)
sns.countplot(x = 'NAME_CONTRACT_TYPE', hue = 'TARGET', data = application_data)
plt.xlabel("Contract Type")

plt.subplot(2,2,2)
percentage_Defaulters('TARGET','NAME_CONTRACT_TYPE', 'Contract Type')

plt.show()
```

#### Following are the observations based on the above graphs:
##### 1. Contract types of Cash Loans are more for both the target values than Revolving Loans
##### 2. Dafaulter percentage of Cash loan clients are more than Revolving loans

```python
plt.figure(figsize = (20,20))
plt.subplot(2,2,1)
sns.countplot(x = 'CODE_GENDER', hue = 'TARGET', data = application_data)
plt.xlabel("Gender")

plt.subplot(2,2,2)
percentage_Defaulters('TARGET','CODE_GENDER', 'GENDER')

plt.show()
```

#### Following are the observations based on the above graphs:
##### 1. Females clients are more in number than males for both the categories.
##### 2. Male clients are more likely to be defaulter than Female.

```python
plt.figure(figsize = (20,20))
plt.subplot(2,2,1)
sns.countplot(x = 'NAME_INCOME_TYPE', hue = 'TARGET', data = application_data)
plt.xlabel("Source of Income")
plt.xticks(rotation=90)

plt.subplot(2,2,2)
percentage_Defaulters('TARGET','NAME_INCOME_TYPE', 'Source of Income')
plt.xticks(rotation=90)

plt.show()
```

#### Following are the observations based on the above graphs:

##### 4. Bank can reduce the loan amount for female applicants who are on maternity leave.

##### 5. Bank should reject the application of young males with lower secondary education and of lower income group and staying with parents or in a rented house, applying for low-range cash contract

##### 6. Banks can target businessmen, students and working class people with academic degree/ higher education to provide loan.
